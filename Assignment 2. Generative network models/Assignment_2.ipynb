### Task 1. Erdos-Renyi model (0 points)

Implement Erdos Renyi model (random graph) — each pair of $n$ nodes are connected with some fixed probability $p$.
Реализовать модель Эрдоса Реньи (случайный граф) — каждая пара из n узлов связана с некоторой фиксированной вероятностью p. 

```python
def erdos_renyi_graph(n, p):
    G = nx.Graph()
    nodes = np.arange(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(random_edges(nodes, p))
    return G
```

Write a function `random_edges` with parameters: `nodes` — np.array of nodes, for example [1, 2, 3, 4, ...] and `p` — probability of connection. The function returns np.array with tuples of the form [(1, 2), (2, 4), ...], where 1-2, 2-4 are edges that should be added in the graph.

Напишите функцию `random_edges` с параметрами: `nodes` — np.массив узлов, например [1, 2, 3, 4, ...] и `p` — вероятность соединения. Функция возвращает np.array с кортежами вида [(1, 2), (2, 4), ...], где 1-2, 2-4 — ребра, которые нужно добавить в граф.

```python
# def random_edges(nodes, p):
# 	    rmatrix = np.random.rand(len(nodes), len(nodes))
# 	    row_ids, col_ids = np.where(rmatrix <= p)
# 	    mask = row_ids > col_ids
# 	    return np.stack([row_ids[mask], col_ids[mask]]).T
#return np.array([(x, y) for x in nodes for y in nodes if x < y and np.random.uniform(0, 1) < p])

def random_edges(nodes, p):
  # YOUR CODE HERE
	    edges = []
	    v, w = 1, -1
	    n = len(nodes)
	    while v < n:
	      r = np.random.uniform(size=1)
	      w = int(w + 1 + np.log(1 - r) / np.log(1 - p))
	      while w >= v and v < n:
	        w -= v
	        v += 1	
	      if v < n:
	        edges.append((v, w))	
	    return np.array(edges)
```

In the Erdos-Renyi model, we can estimate a degree distribution using the binomial distribution. The binomial distribution $B(n, p)$ converges to the Poisson $\text{Pois}(\lambda)$ when the number of samples $n$ tends to infinity with the fixed product $np = \lambda$.

В модели Эрдоса-Реньи мы можем оценить распределение степеней, используя биномиальное распределение. Биномиальное распределение $B(n, p)$ сходится к пуассоновскому $\text{Pois}(\lambda)$, когда число выборок $n$ стремится к бесконечности при фиксированном произведении $np = \lambda$.

In this task we will estimate properties of Binomial and Poisson distributions and compare in what cases it is better to use Binomial.

Write a function `estimate_binomial` that takes a random graph and returns binomial parameters `n` and `p`.

В этой задаче мы оценим свойства биномиального и пуассоновского распределений и сравним, в каких случаях лучше использовать биномиальное.

Напишите функцию  `estimate_binomial`, которая берет случайный граф и возвращает биномиальные параметры `n` and `p`.

```python
def estimate_binomial(G):
    # YOUR CODE HERE
    n = len(G.nodes)
    all_possible_edges =  n * (n - 1) / 2.
    n_edges = len(G.edges)
    p = n_edges / all_possible_edges
    return n, p
```

Посмотрим, как зависит размер мелких компонентов (маленьких — не гигантских) от вероятности вблизи критической точки: p=1/n .

Напишите функцию small_component_size с параметрами n — количество узлов, probabilities — np.массив вероятностей. Функция генерирует случайный граф для каждой вероятности и возвращает np.array средних размеров мелких компонентов.

Подсказка: чтобы найти узлы в компонентах, используйте nx.connected_components(graph) 

```python
def small_component_size(n, probabilities):
    # YOUR CODE HERE
    avr_sc = []
    for p in probabilities:
      G = erdos_renyi_graph(n, p)
      comp = list(nx.connected_components(G))
      s = 0
      n_sm = 0
      for i in range(len(comp)):
        if len(comp[i]) < n/2:
          s+= len(comp[i])
          n_sm +=1
      avr_sc.append(s / n_sm)
    return np.array(avr_sc)
```

```python
def snobbish_network(n, p, q):
  # YOUR CODE HERE
    G = nx.Graph()
    nodes = np.arange(2 * n)
    G.add_nodes_from(nodes)
    labels = {x: 'red' for x in np.arange(n)}
    for i in range(n, 2*n):
      labels[i] = 'blue'
    nx.set_node_attributes(G, labels, 'color')
    arr = np.array([(x, y) for x in nodes for y in nodes if (x < y and np.random.uniform(0, 1) < p and G.nodes[x]['color'] == G.nodes[y]['color']) or \
                    (x < y and np.random.uniform(0, 1) < q and G.nodes[x]['color'] != G.nodes[y]['color'])])
    G.add_edges_from(arr)
    return G
```

Implement Watts-Strogatz model (small-world model) — rewire an edge with probability `p` in a ring lattice with `n` nodes and `k` degree.

Реализовать модель Уоттса-Строгаца (модель маленького мира) — перемонтировать ребро с вероятностью p в кольцевой решетке с n узлами и степенью k. 

```python
def watts_strogatz_graph(n, k, p):
  G = ring_lattice(n, k)
  for node in G.nodes:
    rewire(G, node, k, p)
  return G
```

Write a function `ring_lattice` that returns a regular ring lattice with `n` nodes (0, 1, 2, ..., n-1) and `k` node degree. In a case of an odd node degree, it round it to the nearest smaller even number.

Напишите функцию ring_lattice, которая возвращает обычную кольцевую решетку с n узлами (0, 1, 2, ..., n-1) и k степенью узла. В случае нечетной степени узла она округляется до ближайшего меньшего четного числа. 

```python
def ring_lattice(n, k):
  # YOUR CODE HERE
  G = nx.Graph()
  nodes = list(range(n))
  for i in range(1, int(k/2 + 1)):
    targets = nodes[i:] + nodes[:i]
    G.add_edges_from(zip(nodes, targets))
  return G
```

Write a function `rewire` that takes in input a ring lattice `G`, a `node`, a model parameter `k` and probability `p`. For every right hand side neighbor $i$, the function rewires an edge (`node`, $i$) into a random edge (`node`, $j$) with probability `p` where $i \neq j \neq $ `node`.

*Hints:*

* *To speed up the generation, do not filter nodes to random selection. If a selected node produces an existing edge or a loop, just skip it.*


Напишите функцию rewire, которая принимает на вход кольцевую решетку G, узел, параметр модели k и вероятность p. Для каждого правого соседа i функция заменяет ребро (узел, i) на случайное ребро (узел, j) с вероятностью p, где i≠j≠ узел.

Подсказки:

Почему мы перемонтируем только правые боковые края? Мы хотим гарантировать, что будут перемотаны только нетронутые в предыдущих итерациях ребра. Посмотрите на картинку — мы не могли двигать красные края в предыдущих итерациях

Чтобы ускорить генерацию, не фильтруйте узлы для случайного выбора. Если выбранный узел создает существующее ребро или петлю, просто пропустите его. 

```python
def rewire(G, node, k, p):
    # YOUR CODE HERE
    for i in G.nodes:
      neighbors = (node + np.arange(1, int(k/2)+1)) % n
      for j in neighbors:
        if np.random.random() <= p:
          target = np.random.choice([x for x in range(n) if x != i])
          if not G.has_edge(node, target):
            G.remove_edge(node, j)
            G.add_edge(node, target)
      return G
```

Implement Barabasi-Albert model (preferential attachment model) – a growth process where each new node connects to `m` existing nodes. The higher node degree, the higher probability of the connection. The final number of nodes is `n`.

You start from a star graph with `m + 1` nodes. In each step you create `m` edges between a new node and existing nodes. The probability of connection to the node $i$ is 
$$p(i) = \frac{k_i}{\sum k}$$

Write a function `attach` that attaches a `node` to a graph `G` through `m` edges.

*Hint: Create a list with repeated nodes from a list of edges. For example, $[(1, 2), (2, 3), (2, 4)] \to [1, 2, 2, 3, 2, 4]$. Uniformly select nodes one-by-one. Apply `random.choice` instead of `np.random.choice` to speed up the generation.*


Внедрить модель Барабаси-Альберта (модель предпочтительного присоединения) — процесс роста, при котором каждый новый узел соединяется с m существующими узлами. Чем выше степень узла, тем выше вероятность соединения. Окончательное количество узлов равно n.

Вы начинаете со звездного графа с m + 1 узлами. На каждом шаге вы создаете m ребер между новым узлом и существующими узлами. Вероятность соединения с узлом  $i$ равна 
$$p(i) = \frac{k_i}{\sum k}$$

Напишите функцию attach, которая прикрепляет узел к графу G через m ребер.

Подсказка: Создайте список с повторяющимися узлами из списка ребер. Например,$[(1,2),(2,3),(2,4)]→[1,2,2,3,2,4]$ . Равномерно выберите узлы один за другим. Примените random.choice вместо np.random.choice, чтобы ускорить генерацию. 

```python
'''Do not touch the cell'''
def barabasi_albert_graph(n, m, random_seed=1):
    G = nx.star_graph(m)
    for i in range(1, n - m):
        attach(m + i, G, m)
    return G
```

### Task 9. Degree distribution in preferential attachment (2 points)

Barabasi-Albert graph has a degree distribution of the form

$$P(k) = \frac{2m^2}{k^3}$$

That is Power law with $\alpha = 3$ and $k_\min = m$. 

Check this fact by an experiment — generate a set of Barabasi-Albert graphs and estimate parameters of Power law using MLE

$$\alpha = 1 + n \left[\sum_i \log \frac{k_i}{k_\min} \right]^{-1}$$

where the $k_\min$ is selected by minimal Kolmogorov-Smirnov distance between observed and theoretical distributions.

First, write a function `power_law_cdf` that takes an argument and parameters of the Power law distribution and returns the CDF.


Граф Барабаси-Альберта имеет распределение степеней вида

$$P(k) = \frac{2m^2}{k^3}$$

Это степенной закон с $\alpha = 3$ и $k_\min = m$.

Проверьте этот факт экспериментально — сгенерируйте набор графиков Барабаси-Альберта и оцените параметры степенного закона с помощью MLE.

$$\alpha = 1 + n \left[\sum_i \log \frac{k_i}{k_\min} \right]^{-1}$$

где $k_\min$ выбирается по минимальному расстоянию Колмогорова-Смирнова между наблюдаемым и теоретическим распределениями.

Во-первых, напишите функцию `power_law_cdf`, которая принимает аргумент и параметры распределения по степенному закону и возвращает CDF. 

```python
def power_law_cdf(k, alpha=3.5, k_min=1):
    # YOUR CODE HERE
    C = (alpha - 1) / k_min ** (1 - alpha)
    return 1 + C * k ** (-alpha + 1) / (-alpha + 1)
```

```python
assert power_law_cdf(2, 2, 1) == 0.5
assert power_law_cdf(10, 2, 1) == 0.9
```

Next, write a function `mle_power_law_params` that takes a degree sequence and returns a tuple: the best $\alpha$, the best $k_\min$ w.r.t. MLE

_Hint: use `scipy.stats.kstest` where a theoretical CDF is a `power_law_cdf` function and `args=(alpha, k_min)`_

Затем напишите функцию `mle_power_law_params`, которая принимает последовательность степеней и возвращает кортеж: лучший $\alpha$, лучший $k_\min$ относительно. MLE

_Подсказка: используйте `scipy.stats.kstest`, где теоретический CDF является функцией `power_law_cdf` и `args=(alpha, k_min)`_ 

```python
def mle_power_law_params(degree_sequence):
    # YOUR CODE HERE
  so = 0
  x_min = min(degree_sequence)
  for i in degree_sequence:
    so += np.log(i / x_min)
  a = 1 + len(degree_sequence) * (so) ** -1
  return a, x_min
```

```python
assert mle_power_law_params(np.array([1, 2, 3]))[0] > 0
assert mle_power_law_params(np.array([1, 2, 3]))[1] > 0
```

Write a function `estimate_power_law` that generates Barabasi-Albert graphs with `n` nodes, from `m_min` to `m_max` connections and returns a tuple of np.arrays: $\alpha$ and $k_{\min}$ for each graph.


Напишите функцию `estimate_power_law`, которая генерирует графы Барабаси-Альберта с `n` узлами, от `m_min` до `m_max` соединений и возвращает кортеж из np.arrays: $\alpha$ и $k_{\min}$ для каждого график. 

```python
def estimate_power_law(n, m_min, m_max):
    # YOUR CODE HERE
    alpha = []
    k_min = []
    for i in range(m_min, m_max + 1):
      G = nx.barabasi_albert_graph(n, i)
      degrees = [degree for (node, degree) in G.degree]
      a, k = mle_power_law_params(degrees)
      alpha.append(a)
      k_min.append(k)
    return np.array(alpha), np.array(k_min)
```
